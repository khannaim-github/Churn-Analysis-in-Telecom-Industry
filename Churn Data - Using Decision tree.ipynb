```python
# Import Label Encoder

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# Convert categorical data into Numeric form 

churn[churn.select_dtypes(include=['object']).columns] = churn[churn.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
# Drive data into x and y 
```

```python
churn_x = churn.iloc[ : , 0:21]
churn_y = churn.iloc[ : , -1]
```

```python
# take backup 
churn_x_backup = churn_x
```

```python
# Convert the data frame into  array form

churn_x = np.array(churn_x)
churn_y = np.array(churn_y)
```

```python
# Run the Random Forest Function 

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
# Import the Boruta library

from boruta import BorutaPy
```

```python
# Run the Boruta Function 

boruta_feature_selector = BorutaPy(rfc , max_iter=25 , verbose=2)
boruta_feature_selector
```

```python
# Fit the Boruta function

boruta_feature_selector.fit(churn_x, churn_y)
```

```python
# Run the support function over boruta 

boruta_feature_selector.support_
```

```python
# Take the columns over here

churn_x_backup.columns
```

```python
# Create the Data Frame for feature and Importance 

Feature_Importance = pd.DataFrame({"Feature" : churn_x_backup.columns , "Importance": boruta_feature_selector.support_})
Feature_Importance
```

```python
# Drop unwanted columns 

churn = churn.drop(["gender" , "Partner" , "Dependents" , "CallService" , "MultipleConnections" , "BillingMethod"], axis=1)
```

```python
# Check the head records

churn.head()
```

```python
# Sampling 

from sklearn.model_selection import train_test_split
churn_train, churn_test = train_test_split(churn, test_size = 0.2, random_state = 555)
```

```python
churn_train_x = churn_train.iloc[ : , 0:-1]
churn_train_y = churn_train.iloc[ : , -1]
```

```python
churn_test_x = churn_test.iloc[ : , 0:-1]
churn_test_y = churn_test.iloc[ : , -1]
```

```python
print(churn_train_x.shape)
print(churn_train_y.shape)
print('-----------')
print(churn_test_x.shape)
print(churn_test_y.shape)
```

```python
# Create Decision Tree 

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=5 , min_samples_split= 25 , criterion='gini' ,class_weight='balanced')
```

```python
# Fit the model 

dt.fit(churn_train_x , churn_train_y)
```

```python
# Predict over test data 

pred_dt = dt.predict(churn_test_x)
#pred_dt
```

```python
# Import confusion matrix

from sklearn.metrics import confusion_matrix
```

```python
# Check the confusion matrix

tab_dt = confusion_matrix(pred_dt , churn_test_y)
tab_dt
```

```python
# Check the accuracy 

Acc = tab_dt.diagonal().sum() / tab_dt.sum() * 100
Acc
```

```python
# Check for feature importance over DT

dt.feature_importances_
```

```python
# Take sum of feature importance 

dt.feature_importances_.sum()
```

```python
# Check the length of feature importance

len(dt.feature_importances_)
```

```python
# Create data frame of feature and importance 

Feature_Importance = pd.DataFrame({"Feature" : churn_test_x.columns , "Importance" : dt.feature_importances_})
Feature_Importance.sort_values("Importance" , ascending = False)
```
